```python
# Drop the 4 missing columns
realestate_df.drop(columns=['property_description','property_overview','image_url'],inplace=True)
```

```python
# Fill unknown address 
realestate_df['address'].fillna('Unknown', inplace=True)
```

```python
sns.scatterplot(x='area', y='price', data=realestate_df)
```

```python
# Groupbycity
grouped_by_city = realestate_df.groupby(['city'])
```

```python
# Calculate the total of propertis in each city using
total_propertyareabycity = grouped_by_city['area'].sum()
```

```python
property_density_by_city = total_propertyareabycity / grouped_by_city.size()
```

```python
# How does the property density vary across the country?
# Property density variation across cities
property_density_by_city.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Property density')
plt.show()
```

```python
# Convert the property DataFrame 
property_density_by_city = pd.DataFrame(property_density_by_city)
```

```python
# Merge the dataframe on city
merged_df = realestate_df.merge(property_density_by_city, on='city')
```

```python
# Rename the 0 column
merged_df.rename(columns={0:'property_density'},inplace=True)
```

```python
merged_df.head(5)
```

```python
merged_df.describe()
```

```python
# Does the density of available properties influence the price?
sns.regplot(x='property_density', y='price', data=merged_df)
```

```python
# Relationship between property density and price broken down by city
city_groups = realestate_df.groupby('city')

# Create a scatter plot for each city
for city, data in city_groups:
    # Calculate the property density and price for the current city
    property_density = len(data) / sum(data['area'])
    price = data['price'].mean()
    
    # Create a scatter plot with property density on the x-axis and price on the y-axis
    plt.scatter(property_density, price)
    
    # Add the city name as a label to the scatter plot
    plt.annotate(city, (property_density, price))
    
    # Add axis labels and a title to the scatter plot
    plt.xlabel('Property Density')
    plt.ylabel('Price')
    plt.title('Property Density vs. Price in {}'.format(city))
    
    # Show the scatter plot for the current city
    plt.show()
```

```python
# Calculating property density by locality
locality_propertydensity = merged_df.groupby(['locality'])
```

```python
total_propertyareabylocality = locality_propertydensity['area'].sum()
```

```python
property_density_by_locality = total_propertyareabylocality / locality_propertydensity.size()
```

```python
top_10_localities = property_density_by_locality.nlargest(10)
```

```python
# Top 10 localities with greatest property density
top_10_localities.plot(kind='bar',color='red')
plt.xticks(rotation=90)
plt.xlabel('locality')
plt.ylabel('Property density')
plt.show()
```

```python
#What is the average price for properties in the top 10 localities with the highest property density?
locality_list = ['Kafrul', 'Kaliakair', 'Sagorika Bscic Industrial Area', 'Banasree', 'Rupganj', 'Bashundhara R/A', 'Badda', 'Noakhali Sadar', 'Dhanmondi', 'Zindabazar']
locality_avg_price = merged_df.loc[merged_df['locality'].isin(locality_list)].groupby('locality')['price'].mean()
```

```python
#What is the average price for properties in the top 10 localities with the highest property density?
locality_avg_price.plot(kind='bar',color='red')
plt.xticks(rotation=90)
plt.xlabel('locality')
plt.ylabel('Property density')
plt.show()
```
