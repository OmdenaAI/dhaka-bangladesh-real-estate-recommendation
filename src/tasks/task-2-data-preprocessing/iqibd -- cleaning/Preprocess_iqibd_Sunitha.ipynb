```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
iqibd_df=pd.read_csv('raw/iqibd_properties.csv')
iqibd_df.head()
```

```python
#Get the info on the dataset
iqibd_df.info()
```

The dataset has 261 rows and 33 columns.

### Convert the columns to the required datatypes

```python
#Extracting the numerical value from the area column since the unit is sqft or ft2
iqibd_df['area']=iqibd_df.area.str.split(r'[a-zA-Z]').str[0]
iqibd_df['area'].head()
```

```python
#Converting the datatype to float
iqibd_df['area']=iqibd_df['area'].astype('float')
iqibd_df['area'].dtype
```

```python
#Get the unique values in building_type column
iqibd_df['building_type'].unique()
```

```python
#Set a new column building_nature value as 'Residential' or 'Commercial' based on building_type
iqibd_df["building_nature"] = ""
for index,row in iqibd_df.iterrows():
    if row['building_type']=='Apartment':
        building_nature = 'Residential'
    elif row['building_type']=='Commercial property':
        building_nature = 'Commercial'    
    iqibd_df.at[index,'building_nature'] = building_nature
```

```python
iqibd_df.head()
```

```python
#Setting the country column as 'Bangladesh'
#iqibd_df["country"] = "Bangladesh"
iqibd_df["city"] = iqibd_df.location.str.split(',').str[1]
iqibd_df["address"] = iqibd_df.location.str.split(',').str[0]
#Drop the location column
iqibd_df=iqibd_df.drop(['location'],axis=1)
iqibd_df.head()
```

```python
#Fill the null values with 0 for num_bed_rooms and num_bath_rooms
iqibd_df['num_bed_rooms']=iqibd_df.num_bed_rooms.fillna(0)
iqibd_df['num_bath_rooms']=iqibd_df.num_bath_rooms.fillna(0)
```

```python
#Convert num_bath_rooms and num_bed_rooms dtype as int from float
iqibd_df=iqibd_df.astype({'num_bath_rooms':'int','num_bed_rooms':'int'})
iqibd_df.head()
```

```python
#Extract the price value alone in price column 
iqibd_df['price'] = iqibd_df['price'].str.extract(r'(\d.*)')
#Replace the commas in price string and convert dtype as float
iqibd_df['price']=iqibd_df['price'].str.replace(',','').astype('float')
iqibd_df['price'].head()
```

```python
#Adding new columns property_description,property_overview and setting as null since values not available
iqibd_df["property_description"] = np.nan
iqibd_df["property_overview"] = np.nan
```

```python
#Renaming the Status column as purpose
iqibd_df.rename(columns={'status':'purpose'},inplace=True)
iqibd_df.purpose.value_counts()
```

```python
#Set the value 'For Sale' as 'Sale' as only sale data is available in the site
iqibd_df['purpose']='Sale'
iqibd_df.head()
```

```python
iqibd_df.columns
```

```python
#Fill the amenities column with 'N' for empty values
amenities_list=['air_conditioning','balcony','cctv','central_heating', 'elevator', 'emergency_stairs', 'fitness_center',
                'garden', 'generator', 'government_gas', 'parking','pets_allow', 'reception','security', 'servant_room',
                'swimming_pool', 'tv_cable', 'wifi']
#iqibd_df[amenities_list]=iqibd_df[amenities_list].fillna('N')
iqibd_df.head()
```

```python
#Rename the amenities columns for uniformity
amenity_dict={}
for amenity in amenities_list:
    amenity_dict.update({amenity:amenity+'-amenity'})
amenity_dict    
```

```python
iqibd_df.rename(columns=amenity_dict,inplace=True)
iqibd_df.head()
```

```python
#Replace all the 'Y' values to yes for the amenities columns
amenity_list=list(amenity_dict.values())
iqibd_df[amenity_list]=iqibd_df[amenity_list].replace({'Y':'yes'})
iqibd_df.head()
```

```python
#Renaming the columns again for uniformity
iqibd_df.rename(columns={'balcony-amenity':'balcony-or-terrace-amenity','cctv-amenity':'cctv-security-amenity',
                        'central_heating-amenity':'central-heating-amenity','elevator-amenity':'elevators-in-building-amenity',
                        'garden-amenity':'lawn-or-garden-amenity','parking-amenity':'parking-spaces-amenity',
                        'property_link':'property_url','swimming_pool-amenity':'swimming-pool-amenity',
                        },inplace=True)
```
