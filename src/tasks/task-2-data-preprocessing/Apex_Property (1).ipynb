```python
import pandas as pd 
```

```python
df=pd.read_excel(r"/content/drive/MyDrive/apex_property.xlsx")
df
```

```python

def extract_data(row):
    da=eval(row["data"])
    result=pd.DataFrame(columns=["price","currency","property_size","beds"])
    for item in da: 
        dict = pd.Series({
            "price":item["price"],
            "currency":item["currency"],
            "property_size":item["property_size"],
            "beds":item["beds"]   
        })
        result=result.append(dict,ignore_index=True)
    return result
#new_df=df.apply(extract_data,axis=1)
new_df=pd.concat([extract_data(row) for _,row in df.iterrows()],ignore_index = True)
df=pd.concat([df,new_df],axis=1)
df=df.drop("data",axis=1)
df
```

```python
df1=df.dropna()
```

```python
df1.isnull().sum()
```

```python
# Define the convert_to_thousands function
def convert_to_thousands(value):
    if isinstance(value, str):
        if value.endswith('Crore'):
            return float(value[:-6]) * 10000000
        elif value.endswith('Lakh'):
            return float(value[:-5]) * 100000
        elif value.endswith('Thousand'):
            return float(value[:-9]) * 10000
    return value

# Apply function to "price" column using .loc indexer with both row and column indexers
df1.loc[:,"price"] = df1["price"].apply(convert_to_thousands)
```

```python
df1.loc[:,"beds"] = df1.loc[:,"beds"].str.replace("Beds", "", regex=False)
df1
```

```python
df1.loc[:,"property_size"] = df1.loc[:,"property_size"].str.replace("Sft", "", regex=False)
df1
```

```python
df1["building_nature"]="Residential"
df1["city"]="Dhaka"
df1.rename(columns={'beds': 'num_bed_rooms','property_size':'area[sqft]'}, inplace=True)
df1
```

```python
mask = ~df1["num_bed_rooms"].str.contains(r'Baths')

# Apply the boolean mask to filter out rows from the DataFrame
df1 = df1[mask]

# Display the resulting DataFrame
print(df1)
```

```python
df1.to_csv('apex_property_cleaned.csv', index=False)
```
