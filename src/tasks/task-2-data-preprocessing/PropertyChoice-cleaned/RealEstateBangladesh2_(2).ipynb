##Fill NaN values

```python
valDf = valDf.replace(np.nan, '', regex=True)
valDf.head(2)
```

##Remove Unit from Price

```python
valDf['Price'].unique()
```

```python
valDf['ProcessedPrice']=valDf['Price'].apply(lambda x: x.split("  ")[-1] )
valDf['ProcessedPrice'].unique()
```

```python
valDf['ProcessedPrice']=valDf['ProcessedPrice'].apply(lambda x: x.split(" ")[0] )
valDf['ProcessedPrice']=valDf['ProcessedPrice'].str.replace(",","")
valDf['ProcessedPrice'].unique()
```

```python
valDf["ProcessedPrice"]=valDf["ProcessedPrice"].str.replace("Price", "")
```

##Processing Area Column

```python
valDf['Area'].unique()
```

```python
valDf['Area']=valDf['Area'].str.replace("SqFt","")
valDf['Area']=valDf['Area'].str.replace("sqf","")
valDf['Area']=valDf['Area'].str.replace("Sqft","")
valDf['Area']=valDf['Area'].str.replace("SQF","")
valDf['Area']=valDf['Area'].str.replace("Sq Ft","")
valDf['Area']=valDf['Area'].str.replace("SqF2900","")
valDf['Area']=valDf['Area'].str.replace("sft","")
valDf['Area']=valDf['Area'].apply(lambda x: x.split("/")[0] )
valDf['Area']=valDf['Area'].apply(lambda x: x.split(",")[0] )
valDf['Area'].unique()
```

```python
valDf.drop(valDf.loc[valDf['Area']=="2900Twenty Nine Thousand Tk."].index, inplace = True)
valDf.drop(valDf.loc[valDf['Area']=="Banani"].index, inplace = True)
valDf.drop(valDf.loc[valDf['Area']=="2500twenty five hundred t"].index, inplace = True)
#(valDf[valDf['Area']="2900Twenty Nine Thousand Tk."].index, inplace = True)
valDf=valDf.reset_index()
```

```python
valDf.head()
```

```python
valDf['Area_unt']=valDf['Area'].apply(lambda x: x.split(" ")[-1] )
valDf['Area']=valDf['Area'].apply(lambda x: x.split(" ")[0] )
```

```python
for index, row in valDf.iterrows():
    sample_area = valDf.loc[index, "Area"] 
    unt = valDf.loc[index, "Area_unt"] 
    area = float(sample_area)
    unt=unt.lower()
    if unt=="katha":
        area *= 720
        
    # updating the area of the sample in the dataframe
    valDf.loc[index, "Area"] = area
     
```

```python
valDf['Area'].unique()
```

```python
valDf['Area']=valDf['Area'].astype(float)
```

##Breakdown amenities to columns

```python
valDf['Features'].unique()
```

```python
from pandas.core.apply import validate_func_kwargs
for index, row in valDf.iterrows(): # loop through each sample
    sample_amenities = str(valDf.loc[index, "Features"])
    #print(sample_amenities)
    amenities_dict = eval(sample_amenities)
    
    for i in range(len(amenities_dict)):
      
      column_name = amenities_dict[i]
      
      if column_name not in valDf.columns.to_list():
            valDf[column_name]= np.NaN # Giving NaN as the default value for the column
        
      valDf.loc[index, column_name] = "Yes"
```
